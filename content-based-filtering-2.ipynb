## Import libraries

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.info()
```

```python
credits.info()
```

## Merge datasets with 'title'

```python
movie = movies.merge(credits, on='title')
movie.head()
```

## Filter and keep relevant columns

```python
movie = movie[['movie_id','title','overview','genres','keywords','cast','crew']]
movie.head()
```

```python
movie.shape
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)

movie.shape
```

```python
# Extracting genres['name'] & keywords['name']

import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movie['genres'] = movie['genres'].apply(convert)
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
# Extracting top 3 characters in the cast feature 

def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else: 
            break
    return L

movie['cast'] = movie['cast'].apply(convert2)
```

```python
movie.head()
```

```python
# Extracting crew['director']

def fetch(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movie['crew'] = movie['crew'].apply(fetch)
```

```python
movie.head()
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
# Remove ' ' (spaces) between words

movie["genres"] = movie["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movie["keywords"] = movie["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movie["cast"] = movie["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movie["crew"] = movie["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie["tags"] = movie["overview"] + movie["genres"] + movie["keywords"] + movie["cast"] + movie["crew"]

movie.head()
```

```python
new_movie = movie[['movie_id', 'title', 'tags']]

new_movie.head()
```

```python
new_movie["tags"] = new_movie["tags"].apply(lambda x:" ".join(x))
new_movie["tags"] = new_movie["tags"].apply(lambda x:x.lower())

new_movie.head()
```

```python
# Handle repeated words in tags feature

import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return  ' '.join(y) 

new_movie['tags'] = new_movie['tags'].apply(stem)
```

```python
new_movie.head()
```

```python
# Vectorization

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')

vector = cv.fit_transform(new_movie['tags']).toarray()
```

```python
# Calcualte cosine angle between vectors

from sklearn.metrics.pairwise import cosine_similarity

similar = cosine_similarity(vector)
```

```python
# Function that will return Top 5 movies 

def recommendation(movie):
    movie_index = new_movie[new_movie['title'] == movie].index[0]
    distances = similar[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_movie.iloc[i[0]].title)
```

```python
recommendation('Avatar')
```

```python
recommendation('Spectre')
```

```python
recommendation('John Carter')
```
